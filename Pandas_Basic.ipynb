```python
''' mean & std depends on two things '''

mean_individuals_members=Import_Data_pd[['individuals', 'family_members']].mean()
std_individuals_members=Import_Data_pd[['individuals', 'family_members']].std()

print(mean_individuals_members)
print(std_individuals_members)
```

```python
''' normalize1 and normalize2 '''

def normalize1(item1):
    mean = item1.mean()
    std = item1.std()
    normalize1_col=(item1 - mean) / std
    return normalize1_col

def normalize2(item2):
    max = item2.max()
    min = item2.min()
    normalize2_col=(item2 - min) / (max - min)
    return normalize2_col

normalize_dataframe=Import_Data_pd[['individuals', 'family_members']].agg(
    [normalize1, normalize2]
)

normalize_dataframe.head()
```
